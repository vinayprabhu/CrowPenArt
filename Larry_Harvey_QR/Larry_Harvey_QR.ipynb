# Step -1 : Get Larry's image:

```python

link_image='https://upload.wikimedia.org/wikipedia/commons/5/58/Larry_Harvey_wdydwyd.jpg'
fname_larry=wget.download(link_image)

im = Image.open(fname_larry)
# Help from here: https://stackoverflow.com/questions/16646183/crop-an-image-in-the-centre-using-pil
width, height = im.size   # Get dimensions
print(width, height)
new_width=512
new_height=512
left = (width - new_width)/2
top = (height - new_height)/2
right = (width + new_width)/2
bottom = (height + new_height)/2

im_cropped=im.crop((left, top, right, bottom))

im_cropped.save('larry_512.png')
im_cropped
```
